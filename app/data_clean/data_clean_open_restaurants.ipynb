```python
import numpy as np
import pandas as pd
```

```python
ora = pd.read_csv("data/Open_Restaurant_Applications.csv")
ora
ora_list = ora.columns.values.tolist()
ora_list

ora_new = ora[['Latitude','Longitude','Restaurant Name','Legal Business Name','Postcode', 'Food Service Establishment Permit #',
              'Business Address', 'Borough', 'Time of Submission','Approved for Sidewalk Seating','Approved for Roadway Seating',
              'Seating Interest (Sidewalk/Roadway/Both)']]
ora_new
ora_new.to_csv('data/ora_new.csv')
```

```python
ori = pd.read_csv("data/Open_Restaurants_Inspections.csv")
ori
ori_list = ori.columns.values.tolist()
ori_list

ori_new = ori[['Latitude','Longitude','Postcode','RestaurantName','SeatingChoice','IsSidewayCompliant','IsRoadwayCompliant',
               'SkippedReason','InspectedOn']]
ori_new= ori_new.rename({'RestaurantName': 'Restaurant Name'}, axis='columns')
ori_new
ori_new.to_csv('data/ori_new.csv')
```

```python
scla = pd.read_csv("data/Sidewalk_Caf__Licenses_and_Applications.csv")
scla
scla_list = scla.columns.values.tolist()
scla_list
scla_new = scla[[ 'BUSINESS_NAME', 'ZIP', 'SWC_TYPE', 'SWC_SQ_FT', 'SWC_TABLES', 'SWC_CHAIRS','APP_STATUS',]]
scla_new
scla_new.to_csv('data/scla_new.csv')
```

```python
orr = ora_new.merge(ori_new, how='inner', on=['Restaurant Name','Latitude','Longitude'])
orr = orr.dropna(axis = 0, subset = ['Latitude','Longitude'])
orr
```

```python

orr = orr.sort_values(["Restaurant Name", "InspectedOn"], ascending = (False, True))
orr = orr.drop_duplicates(subset=['Restaurant Name'], keep='first')
orr.to_csv('data/Open_Restaurants.csv')
```
