```python
import numpy as np
import pandas as pd
from csv import DictReader, reader
from datetime import date
```

```python
#read Open_Streets_Locations.csv
raw_df = pd.read_csv('./../output/NYPD_Shooting_Data_Recent.csv')
```

```python
from uszipcode import SearchEngine
search = SearchEngine()
```

```python
for i in range(len(raw_df)):
    result = search.by_coordinates(raw_df.loc[i, 'Latitude'], raw_df.loc[i, 'Longitude'], radius=3)
    raw_df.loc[i, 'zipcode'] = result[0].zipcode
```
