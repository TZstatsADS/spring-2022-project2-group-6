```python
import numpy as np
import pandas as pd
from csv import DictReader, reader
from datetime import date
```

```python
#read Open_Streets_Locations.csv
raw_df = pd.read_csv('Open_Streets_Locations.csv')
```

```python
raw_df
```

```python
# remove the following columns: Object ID, Organization Name, LIONS ID, School
# TODO: will most likely also remove Approved On Street, Approved From Street, Approved To Stree, Approved Monday Open, Approved Monday Close, etc.
raw_df = raw_df.drop(columns=['Object ID', 'Organization Name', 'LIONS ID', 'School', 'Review Status'])
```

```python
raw_df
```

```python
# remove time from start date
newStartDate = raw_df.apply(lambda row: row.apprStartD.split("T")[0], axis=1)
raw_df['StartDate'] = newStartDate
raw_df = raw_df.drop(columns=['apprStartD'])
```

```python
def create_end_date(df):
    new_end_date = df.apply(lambda row: row.apprEndDat.split(" ")[0].split("/")[2] + "-" 
             + row.apprEndDat.split(" ")[0].split("/")[0] + "-" 
             + row.apprEndDat.split(" ")[0].split("/")[1], axis=1)
    return new_end_date
```

```python
# remove time from end date
newEndDate = create_end_date(raw_df)
raw_df['EndDate'] = newEndDate
raw_df = raw_df.drop(columns=['apprEndDat'])
```

```python
days = {}
days["mon"] = "Monday"
days["tue"] = "Tuesday"
days["wed"] = "Wednesday"
days["thu"] = "Thursday"
days["fri"] = "Friday"
days["sat"] = "Saturday"
days["sun"] = "Sunday"
```

```python
def get_days(days_string):
    days_string = days_string.replace(".", ",")
    days_string = days_string.replace(" ", "")
    days_string = days_string.replace("thur", "thu")
    days_list = days_string.split(",")
    new_days_list = []
    for day in days_list:
        new_days_list.append(days[day])
    return new_days_list
```

```python
# get days the street is open as list
newApprovedDays = raw_df.apply(lambda row: get_days(row.apprDaysWe), axis=1)
raw_df['ApprovedDays'] = newApprovedDays
raw_df = raw_df.drop(columns=['apprDaysWe'])
```
