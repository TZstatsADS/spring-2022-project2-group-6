```python
import numpy as np
import pandas as pd
from csv import DictReader, reader
from datetime import date
```

```python
name_to_borough = {}
name_to_borough["Prospect Park West"] = ["Brooklyn"]
name_to_borough["Manhattan Bridge Ped Path"] = ["Manhattan", "Brooklyn"]
name_to_borough["Williamsburg Bridge Bike Path"] = ["Manhattan", "Brooklyn"]
name_to_borough["Ed Koch Queensboro Bridge Shared Path"] = ["Manhattan"]
name_to_borough["Staten Island Ferry"] = ["Staten Island"]
name_to_borough["Pulaski Bridge"] = ["Queens"]
name_to_borough["Kent Ave btw North 8th St and North 9th St"] = ["Brooklyn"]
name_to_borough["Brooklyn Bridge Bike Path"] = ["Manhattan", "Brooklyn"]
name_to_borough["8th Ave at 50th St."] = ["Manhattan"]
name_to_borough["Broadway at 50th St"] = ["Manhattan"]
name_to_borough["Amsterdam Ave at 86th St."] = ["Manhattan"]
name_to_borough["Columbus Ave at 86th St."] = ["Manhattan"]
name_to_borough["Manhattan Bridge Bike Comprehensive"] = ["Manhattan", "Brooklyn"]
```

```python
# get the borough each entry belongs to
borough_list = raw_df["name"].apply(lambda row: name_to_borough[row])
raw_df['borough_list'] = borough_list
```
